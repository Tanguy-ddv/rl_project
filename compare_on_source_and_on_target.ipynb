```python
from session.ACsession import ACSession
import env
import matplotlib.pyplot as plt
import json
```

```python
s = ACSession(env.SOURCE, 'outputs/compare', verbose=100)
s.load_agent()
s.train_and_test_on_other_env(env.TARGET, 10000, 100, 10, seed=42)
```

```python
path = f"outputs/compare/step_{s.get_step()-1}_train_and_test/"
#path = f"outputs/compare/step_10_train_and_test/"

with open(path + "rewards.json", 'r') as f:
    train_rewards = json.load(f)
with open(path + "test_reward.json", 'r') as f:
    test_rewards = json.load(f)
```

```python
train_eps = range(10000)
test_eps = range(0, 10000, 100)
plt.plot(train_eps, train_rewards, label='training')
plt.plot(test_eps, test_rewards, label='testing')
plt.xlabel("episode")
plt.ylabel('reward')
plt.legend()
plt.grid()
plt.show()
```
